### CI and Z value for all three attributes k,rpm,nm  CI= 95.
#### Here 5% splits 2.5 left and 2.5 right in Normal distribution and the rest of data is called 95% CI

```python
df.rename(columns={'Air temperature [K]':"Air_temperature",'Process temperature [K]':"Process_temperature",'Rotational speed [rpm]':"Rotational_speed",'Torque [Nm]':"Torque"},inplace=True)

import scipy.stats as stats
import math
# Get the z-critical value*
z_critical = stats.norm.ppf(q = 0.95)  

# Check the z-critical value
print("z-critical value:")              
print(z_critical) 

# Get the population standard deviation
Air_temperature_stdev = df.Air_temperature.std()  

margin_of_error = z_critical * (Air_temperature_stdev/math.sqrt(len(df)))

sample_mean = df.Air_temperature.mean()

confidence_interval = (sample_mean - margin_of_error,
                       sample_mean + margin_of_error)  

print("Confidence interval:")
print(confidence_interval)
# Get the population standard deviation and CI for Rotational speed
Rotational_speed_stdev = df.Rotational_speed.std()  

margin_of_error = z_critical * (Rotational_speed_stdev/math.sqrt(len(df)))

sample_mean = df.Rotational_speed.mean()

confidence_interval = (sample_mean - margin_of_error,
                       sample_mean + margin_of_error)  

print("Confidence interval:")
print(confidence_interval)
# Get the population standard deviation and CI for Torque
Torque_stdev = df.Torque.std()  

margin_of_error = z_critical * (Torque_stdev/math.sqrt(len(df)))

sample_mean = df.Rotational_speed.mean()

confidence_interval = (sample_mean - margin_of_error,
                       sample_mean + margin_of_error)  

print("Confidence interval:")
print(confidence_interval)
```

```python
# Get the z-critical value*
z_critical = stats.norm.ppf(q = 0.99)  

# Check the z-critical value
print("z-critical value:")              
print(z_critical) 

# Get the population standard deviation
Air_temperature_stdev = df.Air_temperature.std()  

margin_of_error = z_critical * (Air_temperature_stdev/math.sqrt(len(df)))

sample_mean = df.Air_temperature.mean()

confidence_interval = (sample_mean - margin_of_error,
                       sample_mean + margin_of_error)  

print("Confidence interval:")
print(confidence_interval)
# Get the population standard deviation
Rotational_speed_stdev = df.Rotational_speed.std()  

margin_of_error = z_critical * (Rotational_speed_stdev/math.sqrt(len(df)))

sample_mean = df.Rotational_speed.mean()

confidence_interval = (sample_mean - margin_of_error,
                       sample_mean + margin_of_error)  

print("Confidence interval:")
print(confidence_interval)
# Get the population standard deviation
Torque_stdev = df.Torque.std()  

margin_of_error = z_critical * (Torque_stdev/math.sqrt(len(df)))

sample_mean = df.Torque.mean()

confidence_interval = (sample_mean - margin_of_error,
                       sample_mean + margin_of_error)  

print("Confidence interval:")
print(confidence_interval)

```
